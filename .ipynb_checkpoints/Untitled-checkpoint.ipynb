# Metropolis Algorithm for 2D lattice
Here I will gice an example of how to use the files to find the thermodynamic properties of a @D isling model lattice.

```python
from ma2d import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
config_ = {'range': [4, 5.02], 'step': 1, 
           'lattice_size': (10,10), 'error_method': 12,
          'temp': 0, 'num_rounds': 4000}

sim = metropolisAlgorithm2d(config_)
```

```python
sim.run()
#sim.system[1].m
```

```python
fig, ax = plt.subplots(nrows=1,ncols=2, figsize=(32, 32))
plt.tight_layout()

ax = ax.ravel()

for i in range(2):
    ax[i].plot(sim.final_magnetization[i], 'r.')
    #ax[i].plot(np.arange(sim.num_rounds), sim.final_energy[i], 'm')
plt.show()
```

```python
sim.equilib_time = [2000, 2000]
sim.correlation_time()
```
