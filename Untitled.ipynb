# Metropolis Algorithm for 2D lattice
Here I will gice an example of how to use the files to find the thermodynamic properties of a @D isling model lattice.

```python
from ma2d import *
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
config_ = {'range': [1, 5.02], 'step': 0.2, 
           'lattice_size': (60,60), 'error_method': 12,
          'temp': 1, 'num_rounds': 9000}

sim = metropolisAlgorithm2d(config_)
```

```python
sim.run()
#sim.system[1].m
```

```python
fig, ax = plt.subplots(nrows=4,ncols=3, figsize=(32, 32))
plt.tight_layout()

ax = ax.ravel()

for i in range(12):
    ax[i].plot(sim.final_magnetization[i], 'r')
    ax[i].plot(np.arange(sim.num_rounds), sim.final_energy[i], 'm')
plt.show()
```
